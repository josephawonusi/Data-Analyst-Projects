```python
import pandas as pd
import numpy as np

df = pd.read_csv("TestData.csv")
df.tail()
```

```python
# Check for null values
for field in df.columns:
    print(df[field][df[field].isna()])
```

```python
print(df['category_type'].unique())
print(df['transaction_type'].unique())
print(df['sub_category_type'].unique())
```

```python
# Get sum amount
total_df = df.groupby('ext_user_id')['amount'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

# Assuming sum_amount_per_user is already calculated
total_df['amount']= total_df['amount'].astype('float')

# Define the number of bins
num_bins = 3

# Create histogram
plt.hist(total_df, bins=num_bins, color=['skyblue', 'red'], edgecolor='black')

# Add labels and title
plt.xlabel('Sum Amount')
plt.ylabel('Number of Users')
plt.title('Distribution of Sum Amount per User')

# Show plot
plt.show()
```

```python
total_df.to_csv('SumData.csv')
```
